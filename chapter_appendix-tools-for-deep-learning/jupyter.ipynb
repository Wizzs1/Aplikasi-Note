```python
import pandas as pd
import re
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory

# --- Normalisasi (kamus) ---
normalization_map = {
    "gak": "tidak", "ga": "tidak", "gk": "tidak", "nggak": "tidak", "ngga": "tidak",
    "kagak": "tidak", "smpn": "simpan", "udh": "sudah", "udah": "sudah",
    "dpt": "dapat", "tp": "tapi", "yg": "yang"
}

def normalize_text(text, mapping):
    pattern = r"\b(" + "|".join(map(re.escape, mapping.keys())) + r")\b"
    def repl(m):
        w = m.group(0)
        return mapping.get(w.lower(), w)
    return re.sub(pattern, repl, text, flags=re.IGNORECASE)

# --- Step 1: Normalisasi awal ---
text_normalized = normalize_text(text_asli, normalization_map)

# --- Step 2: Tokenisasi ---
tokens = word_tokenize(text_normalized)

# --- Step 3: Case Folding ---
tokens_lower = [t.lower() for t in tokens]

# --- Step 4: Stopword Removal ---
indo_stop = set(stopwords.words("indonesian"))
extra_stop = {"ini", "itu", "dapat", "yang", "dan", "di", "pada"}
indo_stop |= extra_stop
tokens_nostop = [t for t in tokens_lower if t.isalpha() and t not in indo_stop]

# --- Step 5: Stemming ---
stemmer = StemmerFactory().create_stemmer()
tokens_stem = [stemmer.stem(t) for t in tokens_nostop]

# --- Step 6: Normalisasi akhir token ---
tokens_final = [normalization_map.get(t, t) for t in tokens_stem]

# --- Buat tabel hasil ---
df_summary = pd.DataFrame([
    ("Teks Asli", text_asli),
    ("Normalisasi", text_normalized),
    ("Tokenisasi", str(tokens)),
    ("Case Folding", str(tokens_lower)),
    ("Stopword Removal", str(tokens_nostop)),
    ("Stemming", str(tokens_stem)),
    ("Normalisasi Akhir", str(tokens_final)),
    ("Hasil Akhir (gabung)", " ".join(tokens_final))
], columns=["Tahap", "Hasil"])

df_summary
```
